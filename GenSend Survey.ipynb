## 2023 Gensend Survey Results

A survey was sent to 1536 students who were part of the GenSend experience between 2013 and 2023. The survey attempted to measure how influential GenSend has been in helping develop a church planting culture among these students. 64 responses were received (4.2%).

This project looks at the survey results to determine if there is statistical significance in the responses (data features) and to evaluate any correlation between them.

```python
import pandas as pd
import numpy as np
```

Import the dataset

```python
data = pd.read_excel(r"/Users/dbello/Desktop/GenSend Survey/GenSend Survey Results_12.6.23.xlsx")
```

```python
data
```

```python
# Drop unnecessary columns
data = data.drop(columns=['Respondent ID','Start Date', 'End Date', 'IP Address', 'Current Location', 'How can NAMB help you take the next step in church planting?'])
```

One-hot encode all categorical values

```python
data1 = pd.get_dummies(data)
data1
```

```python
# Pull list of 26 column names
data1.columns
```

## Classification by Clustering: no target variables

Let's perform K-Means clustering since this is un unsupervised task (no target labels)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
k=3
kmeans = KMeans(n_clusters=k)
kmeans.fit(data1)
y_pred = kmeans.predict(data1) # y_pred in this case is the same as the classification labels (kmeans.labels_)
```

```python
y_pred # This gives us an array of the labels (indices) used by the algorithm
```

```python
kmeans.cluster_centers_ # Three centroids of 26 dimensions each (corresponding to 26 features)
```

```python
kmeans.transform(data1) # Each line is a data feature and each column (0, 1, 2) the distance to index of a centroid
```

```python
data1['indices'] = y_pred.tolist() # add array of centroid indices to df as a new column
data1
```

```python
data1.to_excel('output.xlsx', index=False) # export df to excel for visual analysis
```

We will now remove the "indices" column from data1 to perform statistical analysis using "Involved in Church Planting after GenSend_Yes" as target variable.

```python
data1 = data1.drop(columns=['indices'])
```

Let's use a Random Forest Regressor to find the most significant features in the dataset.

```python
# Let's load the packages
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
# Split dataset for training and testing.

X = data1.drop(['Involved in Church Planting after GenSend_Yes', 'Involved in Church Planting after GenSend_No'], axis=1)
y = data1['Involved in Church Planting after GenSend_Yes']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

Find optimal features for Random Forest Regressor

```python
# Import the GridSearchCV class from the sklearn.model_selection library.
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'max_depth': [7,8,9], 
              'n_estimators': [150,200], 
              'min_samples_split': [9,10,11]}

RF_clf_grid_search_cv = GridSearchCV(RandomForestRegressor(random_state = 42), 
                              param_grid=param_grid, verbose=1, cv=10)

RF_clf_grid_search_cv.fit(X_train, y_train)

print("The best parameters are: ", RF_clf_grid_search_cv.best_params_)
```

```python
# Fit the Random Forest Regressor with 200 Decision Trees and max_depth=2

rf = RandomForestRegressor(n_estimators=200, max_depth=7, min_samples_split=11, random_state=42)
rf.fit(X_train, y_train)
```

```python
# Let's create a table with the feature importance values and their corresponding names
# First, we convert each array to a dataframe

df1 = pd.DataFrame(rf.feature_importances_)
df2 = pd.DataFrame(data1.columns)

#Then, we concatanate the resuting df's and add names to the columns

feat_impt = pd.concat([df2, df1], axis=1)
feat_impt.columns =['Feature_Name', 'Value']
feat_impt = feat_impt.sort_values(["Value"], ascending=False)
feat_impt.head(10)
```
